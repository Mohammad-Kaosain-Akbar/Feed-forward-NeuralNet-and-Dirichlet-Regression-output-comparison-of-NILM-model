# Here 'Hour' variable was considered while performing the Dirichlet Regression : (Y ~ Total_Value + Hour)

```python
import pandas as pd
import numpy as np
```

# Output of the model trained by Dirichlet Regression

```python
dout = pd.read_csv('DirichletOutput.csv')
dout
```

# Original Test Data

```python
dtest = pd.read_csv('DirichletTest.csv')
dtest
```

# Output of the model trained by Feed Forward Neural Network

```python
dNout = pd.read_csv('NNoutput.csv')
dNout
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import Image, display

mpl.rcParams['figure.figsize'] = (20,15)
mpl.rcParams['axes.grid'] = False
```

# Analyzing Outputs for VENTILATION

```python
vent_out = dout['V1']
vent_out = np.asarray(vent_out).reshape(-1)
vent_test = dtest['Y.Ventilation']
vent_test = np.asarray(vent_test).reshape(-1)
vent_Nout = dNout['Ventilation']
vent_Nout = np.asarray(vent_Nout).reshape(-1)
```

```python
plt.plot(vent_out, label='Dirichlet') #plot the loss
plt.plot(vent_test, label='True Value') #plot the validation loss
plt.plot(vent_Nout, label='Neural Network') #plot the validation loss
plt.legend(['Dirichlet Output', 'Original Value', 'Neural Network Output'], loc='upper right')
plt.title('Graph for Ventillation Prediction')
```
